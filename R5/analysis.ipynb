```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. copy rhianna on roses
names = set()

prices = pd.read_csv('data/r1_prices.csv')
trades = pd.read_csv('data/r1_trades.csv')

product = 'STARFRUIT'
person = 'Rhianna'
volFilter = 0

roses = prices[prices['product'] == product]['mid_price']
roses.reset_index(drop=True, inplace=True)
roses.plot()

for i in range(len(trades)):
    names.add(trades['buyer'][i])
    names.add(trades['seller'][i])

print(names)


prod_trades = trades[trades['symbol'] == product]
person_buys = prod_trades[(prod_trades['buyer'] == person) & (prod_trades['quantity'] >= volFilter)]
person_sells = prod_trades[(prod_trades['seller'] == person) & (prod_trades['quantity'] >= volFilter)]
buy_times = person_buys['timestamp'].to_list()
sell_times = person_sells['timestamp'].to_list()

roses_list = roses.to_list()
plt.scatter(buy_times, [roses_list[t] for t in buy_times], color='g', marker='^', s=100, zorder=20)
plt.scatter(sell_times, [roses_list[t] for t in sell_times], color='r', marker='v', s=100, zorder=20)
```
