```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('r2d-1.csv', sep=';')
df1 = df1[df1['product'] == 'COCONUTS']['mid_price']
df1 = df1.reset_index(drop=True)
df2 = pd.read_csv('r4_prices.csv')
df2 = df2[df2['product'] == 'COCONUT']['mid_price']
df2 = df2.reset_index(drop=True)[0:10000] - 2000
df1 = df1.rolling(window=10).mean()[100:200]
df2 = df2.rolling(window=10).mean()[100:200]

df1.diff().plot()
df2.diff().plot()
```
