```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('all.csv')

choc = df[df['product'] == 'CHOCOLATE']['mid_price'].reset_index(drop=True)
strawb = df[df['product'] == 'STRAWBERRIES']['mid_price'].reset_index(drop=True)
roses = df[df['product'] == 'ROSES']['mid_price'].reset_index(drop=True)
basket = df[df['product'] == 'GIFT_BASKET']['mid_price'].reset_index(drop=True)

df2 = pd.DataFrame()
df2['combined'] = (6 * strawb + 4 * choc + roses + 381).shift(0)
df2['basket'] = basket

df2[20000:30000].plot()
```
