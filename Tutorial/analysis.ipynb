```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

product = 'STARFRUIT'

# Load the data
data = pd.read_csv('data.csv', delimiter=';')
data = data[data['product']== product]
# keep only column timestamp and mid_price
data = data[['timestamp', 'mid_price']]

# plot
plt.plot(data['timestamp'], data['mid_price'])
plt.show()

```
