```python
def p(x):
  return 0.0002*(x-900)

def scale(x, avg):
  return (1000-avg)/(1000-x)

```

```python

for avg in range(900, 1001):
  lis = []
  for i in range(900, 1001):
    for j in range(i+1, 1001):
      profit = 0
      for k in range(900, i):
        profit += p(k) * (1000-i)
      if j < avg:
        for k in range(i, j):
          profit += p(k) * (1000-avg)  
      else:
        for k in range(i, j):
          profit += p(k) * (1000-j)
      lis.append((profit, i, j))
  lis.sort()
  print(avg, lis[-1], lis[-2], lis[-3])
```
